```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
survey_data = pd.read_excel("<LINK_TO_SPREADSHEET>")
survey_data
```

```python
average_scores = survey_data.iloc[:, [2, 4, 6, 8, 10, 12, 14, 16, 17, 18, 19, 24]].astype(float)
medians = pd.DataFrame(average_scores.median(), columns=["Median Score"])
orig_scores = average_scores
average_scores = pd.DataFrame(average_scores.mean(), columns=["Average Score"]) 
average_scores
```

```python
orig_scores
```

```python
medians
```

```python
confidence = average_scores.iloc[7:9]
ease = average_scores.iloc[9:11]
confidence, confidence.iloc[0,0]
```

```python
def autolabel(ax, rects):
    """Attach a text label above each bar in *rects*, displaying its height in small boxes."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(round(height, 2)),
                    xy=(rect.get_x() + rect.get_width() / 2, height / 2),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
```

```python

swirl = [confidence.iloc[0, 0], ease.iloc[1, 0]]
warninginspector = [confidence.iloc[1, 0], ease.iloc[0, 0]]
labels = ['Confidence', 'Ease of Use']

# calculate the confidence intervals
swirl_std = [orig_scores.iloc[7].std(), orig_scores.iloc[10].std()]
print(orig_scores.iloc[7].std())
warninginspector_std = [orig_scores.iloc[8].std(), orig_scores.iloc[9].std()]
print(warninginspector_std)

x = np.arange(len(labels))
print(x)
print(swirl)

width = 0.35

fig, ax = plt.subplots()

rects1 = ax.bar(x - width/2, swirl, width, label='SWIRL', color='lightblue', yerr=swirl_std, capsize=5)
rects2 = ax.bar(x + width/2, warninginspector, width, label='Baseline', color='lightseagreen', yerr=warninginspector_std, capsize=5)

autolabel(ax, rects1)
autolabel(ax, rects2)

ax.yaxis.grid(True, linestyle='--', linewidth=0.5, color='darkgray', alpha=0.5)
ax.set_ylabel('Score')
ax.set_title('Comparison of Mean Confidence and Ease Scores')
ax.set_xticks(x)
plt.xticks(rotation=0)
ax.set_xticklabels(labels)
ax.legend()
ax.set_ylim(1, 5)

fig.tight_layout()

plt.show()
fig.savefig('plot_combined_superlarge.pdf')
```

```python
post_task_data = pd.read_excel("<LINK_TO_SPREADSHEET>")
post_task_data
```

```python
post_task_data_swirl = post_task_data[post_task_data.iloc[:, 3].str.contains("SWIRL")]
post_task_data_warninginspector = post_task_data[post_task_data.iloc[:, 3].str.contains("WarningInspector")]

post_task_data_swirl = post_task_data_swirl.iloc[:, 4:9]
post_task_data_warninginspector = post_task_data_warninginspector.iloc[:, 4:9]
post_task_data_swirl
```

```python
labels = ['Mental Demand', 'Hurry', 'Performance', 'Effort', 'Frustration']

# Boxplot for SWIRL 
fig, [ax1, ax2] = plt.subplots(1, 2, figsize=(15, 5))
# set color of box to lightblue
plt1 = ax1.boxplot(post_task_data_swirl, patch_artist=True, boxprops=dict(facecolor='lightblue'), showmeans=True, meanline=True, meanprops=dict(color='red'))
ax1.set_title('SWIRL')
ax1.set_xticklabels(labels)
ax1.set_ylabel('Score')
ax1.yaxis.grid(True, linestyle='--', linewidth=0.5, color='darkgray', alpha=0.5)
ax1.legend([plt1['medians'][0], plt1['means'][0]], ['median', 'mean'])


# Boxplot for WarningInspector
plt2 = ax2.boxplot(post_task_data_warninginspector, patch_artist=True, boxprops=dict(facecolor='lightseagreen'), showmeans=True, meanline=True, meanprops=dict(color='red'))
ax2.set_title('Baseline')
ax2.set_xticklabels(labels)
ax2.yaxis.grid(True, linestyle='--', linewidth=0.5, color='darkgray', alpha=0.5)
ax2.set_ylabel('Score')
ax2.legend([plt2['medians'][0], plt2['means'][0]], ['median', 'mean'])

plt.show()
fig.savefig('boxplot_combined_superlarge.pdf', bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

# Assuming post_task_data_swirl and post_task_data_warninginspector are defined
labels = ['Mental Demand', 'Hurry', 'Performance', 'Effort', 'Frustration']

positions_swirl = [1, 3.5, 6, 8.5, 11]  # Adjusted odd positions for SWIRL
positions_warninginspector = [2, 4.5, 7, 9.5, 12]  # Adjusted even positions for WarningInspector

fig, ax = plt.subplots(figsize=(10, 5))

plt1 = ax.boxplot(post_task_data_swirl, positions=positions_swirl, patch_artist=True,
                  boxprops=dict(facecolor='lightblue'), showmeans=True, meanline=True,
                  meanprops=dict(color='red', linestyle='dotted'), widths=0.4)

plt2 = ax.boxplot(post_task_data_warninginspector, positions=positions_warninginspector, patch_artist=True,
                  boxprops=dict(facecolor='lightseagreen'), showmeans=True, meanline=True,
                  meanprops=dict(color='red', linestyle='dotted'), widths=0.4)

ax.set_title('Task Performance Comparison')
ax.set_ylabel('Score')
ax.set_xticks([(a+b)/2 for a, b in zip(positions_swirl, positions_warninginspector)])
ax.set_xticklabels(labels)

ax.yaxis.grid(True, linestyle='--', linewidth=0.5, color='darkgray', alpha=0.5)

# Create legend entries
swirl_patch = mlines.Line2D([], [], color='lightblue', marker='s', linestyle='None',
                            markersize=10, label='SWIRL')
warning_inspector_patch = mlines.Line2D([], [], color='lightseagreen', marker='s', linestyle='None',
                                        markersize=10, label='Baseline')
mean_line = mlines.Line2D([], [], color='red', linestyle='dotted', markersize=15, label='Mean')
median_line = mlines.Line2D([], [], color='orange', linestyle='-', label='Median')

# Add legend to the plot
ax.legend(handles=[swirl_patch, warning_inspector_patch, mean_line, median_line], loc='upper right')

plt.tight_layout()
plt.show()

# Save the figure
fig.savefig('boxplot_combined_superlarge_means.pdf', bbox_inches='tight')
```
